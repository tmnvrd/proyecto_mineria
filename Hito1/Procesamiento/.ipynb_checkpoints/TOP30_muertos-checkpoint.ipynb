### --------------------------------------------------------------------------------------------------------------------------------------------

### VISUALIZACIÓN DE DATASETS

#### 1) time_series_covid_19_confirmed

```python
#Si al exportar los datos del git no tienen la carpeta COVID-19, correr path como: csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv
datos_global_confirmed = pd.read_csv('Data/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv',  sep=',')
datos_global_confirmed.head()
```

#### 2) time_series_covid_19_deaths

```python
datos_global_deaths = pd.read_csv('Data/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv',  sep=',')
datos_global_deaths.head()
```

#### 3) time_series_covid_19_recovered

```python
datos_global_recovered = pd.read_csv('Data/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv',  sep=',')
datos_global_recovered.head()
```

#### 4) countriesWorld

```python
datos_countriesWorld = pd.read_excel('Data/countriesWorld_2.xlsx',  sheet_name='final')
datos_countriesWorld.head()
```

```python
datos_countriesWorld.shape
```

### --------------------------------------------------------------------------------------------------------------------------------------------

### UNIÓN DE DATASETS

```python
#Nombres de las columnas del nuevo dataset, union de series de tiempo + variables del pais
lista =list(datos_global_deaths.columns)[1:]
lista_countries =list(datos_countriesWorld.columns)
for i in range(len(lista_countries)-1):
    lista.append(lista_countries[i+1])
```

```python
#Nuevo dataset, donde los tres primeros datos corresponden al nombre_pais, lat y long
# de la columna 4 al 95 son los muertos del 22/1/20 al 22/4/20
# de la columna 96 corresponde a las variables de cada pais, ej: GDP, Population
full_death = pd.DataFrame(columns=lista)

myset = set(datos_global_deaths['Country/Region'])
i=0
for name in list(myset):
    full_death.loc[i,:3] = datos_global_deaths[(datos_global_deaths['Country/Region']==name)].iloc[0,1:4]
    full_death.loc[i,3:95] = (datos_global_deaths[(datos_global_deaths['Country/Region']==name)].iloc[:,4:]).sum(axis=0)
    full_death.loc[i,95:] = datos_countriesWorld[(datos_countriesWorld['Country/Region']==name)].iloc[0,1:]
    i += 1
    
```

```python
# Visualizacion de datos para Chile
full_death[(full_death['Country/Region']=='Chile')]
```

### --------------------------------------------------------------------------------------------------------------------------------------------

### ELIMINACIÓN DATOS NaN O VACÍOS

```python
# BORRAR LOS NA
full_death.dropna(axis=0, inplace=True)
full_death = full_death.reset_index(drop=True)
```

```python
#Dimension del dataframe resultante
full_death.shape
```

```python
#Reconfirmacion de datos NaN
full_death.isnull().values.any()
```

### --------------------------------------------------------------------------------------------------------------------------------------------

### NORMALIZACION DE CASOS MUERTOS

```python
#Se ordenan los paises alfabeticamente
full_death = full_death.sort_values('Country/Region')
```

```python
full_death.iloc[143]
```

```python
#Vector que contiene los porcentajes de contagiados 
porcentajeMuertos = pd.DataFrame((100*full_death.iloc[:,94])/full_death.iloc[:,100],columns=['Porcentaje'])
porcentajeMuertos
```

```python
#Dataframe de los muertos en porcentaje, con nombre_pais + porcentaje
muertosNormalizados = pd.concat([full_death.iloc[:,0],porcentajeMuertos], axis=1)
muertosNormalizados
```

### --------------------------------------------------------------------------------------------------------------------------------------------

### TOP 30 CASOS MUERTOS 

```python
#Extraccion de los primeros 30 paises mas muertos en porcentaje
top30_norm = muertosNormalizados.sort_values('Porcentaje',ascending=False).iloc[:29,:]
top30_norm
```

```python
#Extraccion de los 30 paises con mas muertos, sin normalizar
muertosSinNormalizados = pd.concat([full_death.iloc[:,0],full_death.iloc[:,94]], axis=1)
top30_NO_norm = muertosSinNormalizados.sort_values('4/22/20',ascending=False).iloc[:29,:]
top30_NO_norm
```

### CURVAS DE TIEMPO: TOP 30 MUERTOS NORMALIZADOS

```python
#Lista con los 10 paises con mas contagioados segun porcentaje
top30_norm_lista = top30_norm['Country/Region'].unique()
top30_norm_lista[:10]
```

```python
#Nombres de las columnas del nuevo dataset, nombre_pais + union de series de tiempo
lista_top30_norm =['Country/Region']
lista_fechas =list(datos_global_deaths.columns)[4:]
for i in range(len(lista_fechas)):
    lista_top30_norm.append(lista_fechas[i])
```

```python
#Los datos del TOP30_normalizados se separan en 2 dataframes por facilidades de plot
# full_death_top30_norm10: Primeros 10 paises del TOP30 normalizados
# full_death_top30_norm20: 20 paises restantes del TOP30 normalizados

#Cada dataframe contiene el nombre del pais en la primera columna
# de la columna 2 al 95 son los muertos del 22/1/20 al 22/4/20

full_death_top30_norm10 = pd.DataFrame(columns=lista_top30_norm)
full_death_top30_norm20 = pd.DataFrame(columns=lista_top30_norm)

myset_death10 = set(top30_norm_lista[:10])
myset_death20 = set(top30_norm_lista[10:])
i=0
for name in list(myset_death10):
    full_death_top30_norm10.loc[i,:1] = name
    full_death_top30_norm10.loc[i,1:] = (100*(datos_global_deaths[(datos_global_deaths['Country/Region']==name)].iloc[:,4:]).sum(axis=0))/datos_countriesWorld[(datos_countriesWorld['Country/Region']==name)].iloc[0,6]
    i += 1
for name in list(myset_death20):    
    full_death_top30_norm20.loc[i,:1] = name
    full_death_top30_norm20.loc[i,1:] = (100*(datos_global_deaths[(datos_global_deaths['Country/Region']==name)].iloc[:,4:]).sum(axis=0))/datos_countriesWorld[(datos_countriesWorld['Country/Region']==name)].iloc[0,6]
    i += 1
```

```python
#se trasponen los datos para que los ejes coicidan en el plot
df10 = full_death_top30_norm10.T
header10 = df10.iloc[0]
df10 = df10[1:]
df10.columns = header10
df10
```

```python
df20 = full_death_top30_norm20.T
header20 = df20.iloc[0]
df20 = df20[1:]
df20.columns = header20
```

```python
# Plot del TOP30 normalizado [%], se destacan los primero 10 paises
%matplotlib notebook
ax = df10.plot(style='*')
df20.plot(ax = ax, style='-')
plt.xlabel('Fecha (22/01/2020 - 20/04/2020)')
plt.ylabel('N° Muertos [%]')
plt.title('Casos muertos a la fecha, para TOP30 normalizado')
plt.legend(df10.columns)
plt.show()
```

```python
#Scatter plots para confirmados con normalizar
%matplotlib notebook

#Poblacion
plt.figure(1)
k = 0
for i in new2['Country/Region'].values:
        jei=full_death[(full_death['Country/Region']==i)]['Population'].values
        if(k<=5):
            #print(i)
            plt.plot(jei,float(new2[new2['Country/Region']==i]['Porcentaje']),'*', label=i)
        else:
            plt.plot(jei,float(new2[new2['Country/Region']==i]['Porcentaje']),'b.') 
        k += 1
plt.title('Compración de N° de muertos según Población')
plt.xlabel("Población")
plt.ylabel("Casos Muertos[%]")
plt.legend()
plt.show()

#Ingreso Per Capita
plt.figure(2)
k = 0
for i in new2['Country/Region'].values:
        jei=full_death[(full_death['Country/Region']==i)]['GDP ($ per capita)'].values
#        if i=="Chile":
#            plt.plot(jei,float(new2[new2['Country/Region']==i]['Porcentaje']),'*', label="Chile")
        if(k<=5):
            plt.plot(jei,float(new2[new2['Country/Region']==i]['Porcentaje']),'*', label=i)
        else:
            plt.plot(jei,float(new2[new2['Country/Region']==i]['Porcentaje']),'b.')
        k += 1
plt.title('Comparación de N° de muertos según Ingreso per capita')
plt.xlabel("GPD [USD/capita]")
plt.ylabel("Casos Muertos [%]")
plt.legend()
plt.show()

#Alfabetización
plt.figure(3)
k = 0
for i in new2['Country/Region'].values:
        jei=full_death[(full_death['Country/Region']==i)]['Literacy (%)'].values
        if(k<=5):
            plt.plot(jei,float(new2[new2['Country/Region']==i]['Porcentaje']),'*', label=i)
        else:
            plt.plot(jei,float(new2[new2['Country/Region']==i]['Porcentaje']),'b.')
        k += 1
plt.title('Comparación de N° de muertos según Alfabetización')
plt.xlabel("Alfabetización [%]")
plt.ylabel("Casos Muertos [%]")
plt.legend()
plt.show()

#Densidad Poblacional
plt.figure(4)
k=0
for i in new2['Country/Region'].values:
        jei=full_death[(full_death['Country/Region']==i)]['Pop. Density (per sq. mi.)'].values
        if i=="Monaco" or i=="Singapore" or i=="Bahamas" or i=="Bahrain" or i=="Bangladesh" or i=="Malta" or i=="Maldives":
            pass 
        else:        
            if(k<=5):
                plt.plot(jei,float(new2[new2['Country/Region']==i]['Porcentaje']),'*', label=i)
            else:
                plt.plot(jei,float(new2[new2['Country/Region']==i]['Porcentaje']),'b.')
            k += 1
plt.title('Comparación de N° de muertos según Densidad Poblacional')
plt.xlabel("Densidad Poblacional [sq. mi]")
plt.ylabel("Casos Muertos [%]")
plt.legend()
plt.savefig('Densidad_poblacional_vs_muertos_n.png')
plt.show()

#Tasa de Mortalidad
plt.figure(5)
k=0
for i in new2['Country/Region'].values:
        jei=full_death[(full_death['Country/Region']==i)]['Deathrate'].values
        if(k<=5):
            plt.plot(jei,float(new2[new2['Country/Region']==i]['Porcentaje']),'*', label=i)
        else:
            plt.plot(jei,float(new2[new2['Country/Region']==i]['Porcentaje']),'b.')
        k += 1
plt.title('Comparación de N° de muertos según Tasa de mortalidad')
plt.xlabel("Tasa Mortalidad [%]")
plt.ylabel("Casos Muertos [%]")
plt.legend()
plt.show()


#Tasa de Natalidad
plt.figure(6)
k=0
for i in new2['Country/Region'].values:
        jei=full_death[(full_death['Country/Region']==i)]['Birthrate'].values
        if(k<=5):
            plt.plot(jei,float(new2[new2['Country/Region']==i]['Porcentaje']),'*', label=i)
        else:
            plt.plot(jei,float(new2[new2['Country/Region']==i]['Porcentaje']),'b.')
        k += 1
plt.title('Comparación de N° de muertos según Tasa de natalidad')
plt.xlabel("Tasa Natalidad [%]")
plt.ylabel("Casos Muertos [%]")
plt.legend()
plt.show()
```
