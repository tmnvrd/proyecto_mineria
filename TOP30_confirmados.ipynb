### --------------------------------------------------------------------------------------------------------------------------------------------

### UNIÓN DE DATASETS

```python
#Nombres de las columnas del nuevo dataset, union de series de tiempo + variables del pais
lista =list(datos_global_confirmed.columns)[1:]
lista_countries =list(datos_countriesWorld.columns)
for i in range(len(lista_countries)-1):
    lista.append(lista_countries[i+1])
```

```python
#Nuevo dataset, donde los tres primeros datos corresponden al nombre_pais, lat y long
# de la columna 4 al 95 son los muertos del 22/1/20 al 22/4/20
# de la columna 96 corresponde a las variables de cada pais, ej: GDP, Population
full_confirmed = pd.DataFrame(columns=lista)

myset = set(datos_global_confirmed['Country/Region'])
i=0
for name in list(myset):
    full_confirmed.loc[i,:3] = datos_global_confirmed[(datos_global_confirmed['Country/Region']==name)].iloc[0,1:4]
    full_confirmed.loc[i,3:95] = (datos_global_confirmed[(datos_global_confirmed['Country/Region']==name)].iloc[:,4:]).sum(axis=0)
    full_confirmed.loc[i,95:] = datos_countriesWorld[(datos_countriesWorld['Country/Region']==name)].iloc[0,1:]
    i += 1
    
```

```python
# Visualizacion de datos para Chile
full_confirmed[(full_confirmed['Country/Region']=='Chile')]
```

### --------------------------------------------------------------------------------------------------------------------------------------------

### ELIMINACIÓN DATOS NaN O VACÍOS

```python
# BORRAR LOS NA
full_confirmed.dropna(axis=0, inplace=True)
full_confirmed = full_confirmed.reset_index(drop=True)
```

```python
#Dimension del dataframe resultante
full_confirmed.shape
```

```python
#Confirmacion de datos NaN
full_confirmed.isnull().values.any()
```

### --------------------------------------------------------------------------------------------------------------------------------------------

### NORMALIZACION DE CASOS CONFIRMADOS

```python
#Se ordenan los paises alfabeticamente
full_confirmed = full_confirmed.sort_values('Country/Region')
```

```python
full_confirmed.iloc[143]
```

```python
#Vector que contiene los porcentajes de contagiados 
porcentajeConfirmados = pd.DataFrame((100*full_confirmed.iloc[:,94])/full_confirmed.iloc[:,100],columns=['Porcentaje'])
porcentajeConfirmados
```

```python
#Dataframe de los confirmados en porcentaje, con nombre_pais + porcentaje
confirmadosNormalizados = pd.concat([full_confirmed.iloc[:,0],porcentajeConfirmados], axis=1)
confirmadosNormalizados
```

### --------------------------------------------------------------------------------------------------------------------------------------------

### TOP 30 CASOS CONFIRMADOS 

```python
#Extraccion de los primeros 30 paises mas contagiados en porcentaje
top30_norm = confirmadosNormalizados.sort_values('Porcentaje',ascending=False).iloc[:29,:]
top30_norm
```

```python
#Extraccion de los 30 paises con mas confirmados, sin normalizar
confirmadosSinNormalizados = pd.concat([full_confirmed.iloc[:,0],full_confirmed.iloc[:,94]], axis=1)
top30_NO_norm = confirmadosSinNormalizados.sort_values('4/22/20',ascending=False).iloc[:29,:]
top30_NO_norm
```

### --------------------------------------------------------------------------------------------------------------------------------------------

### CURVAS DE TIEMPO: TOP 30 CONFIRMADOS NORMALIZADOS

```python
#Lista con los 10 paises con mas contagiados segun porcentaje
top30_norm_lista = top30_norm['Country/Region'].unique()
top30_norm_lista[:10]
```

```python
#Nombres de las columnas del nuevo dataset, nombre_pais + union de series de tiempo
lista_top30_norm =['Country/Region']
lista_fechas =list(datos_global_confirmed.columns)[4:]
for i in range(len(lista_fechas)):
    lista_top30_norm.append(lista_fechas[i])
```

```python
#Los datos del TOP30_normalizados se separan en 2 dataframes por facilidades de plot
# full_confirmed_top30_norm10: Primeros 10 paises del TOP30 normalizados
# full_confirmed_top30_norm20: 20 paises restantes del TOP30 normalizados

#Cada dataframe contiene el nombre del pais en la primera columna
# de la columna 2 al 95 son los muertos del 22/1/20 al 22/4/20

full_confirmed_top30_norm10 = pd.DataFrame(columns=lista_top30_norm)
full_confirmed_top30_norm20 = pd.DataFrame(columns=lista_top30_norm)

myset_confirmed10 = set(top30_norm_lista[:10])
myset_confirmed20 = set(top30_norm_lista[10:])
i=0
for name in list(myset_confirmed10):
    full_confirmed_top30_norm10.loc[i,:1] = name
    full_confirmed_top30_norm10.loc[i,1:] = (100*(datos_global_confirmed[(datos_global_confirmed['Country/Region']==name)].iloc[:,4:]).sum(axis=0))/datos_countriesWorld[(datos_countriesWorld['Country/Region']==name)].iloc[0,6]
    i += 1
for name in list(myset_confirmed20):    
    full_confirmed_top30_norm20.loc[i,:1] = name
    full_confirmed_top30_norm20.loc[i,1:] = (100*(datos_global_confirmed[(datos_global_confirmed['Country/Region']==name)].iloc[:,4:]).sum(axis=0))/datos_countriesWorld[(datos_countriesWorld['Country/Region']==name)].iloc[0,6]
    i += 1
```

```python
#se trasponen los datos para que los ejes coicidan en el plot
df10 = full_confirmed_top30_norm10.T
header10 = df10.iloc[0]
df10 = df10[1:]
df10.columns = header10
df10
```

```python
df20 = full_confirmed_top30_norm20.T
header20 = df20.iloc[0]
df20 = df20[1:]
df20.columns = header20
```

```python
# Plot del TOP30 normalizado [%], se destacan los primero 10 paises
%matplotlib notebook
ax = df10.plot(style='*')
df20.plot(ax = ax, style='-')
plt.xlabel('Fecha (22/01/2020 - 20/04/2020)')
plt.ylabel('N° Confirmados [%]')
plt.title('Casos confirmados a la fecha, para TOP30 normalizado')
plt.legend(df10.columns)
plt.show()
```
