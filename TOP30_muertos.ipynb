### VISUALIZACIÓN DE DATASETS

#### 1) time_series_covid_19_confirmed

```python
datos_global_confirmed = pd.read_csv('COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv',  sep=',')
datos_global_confirmed.head()
```

#### 2) time_series_covid_19_deaths

```python
datos_global_deaths = pd.read_csv('COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv',  sep=',')
datos_global_deaths.head()
```

#### 3) time_series_covid_19_recovered

```python
datos_global_recovered = pd.read_csv('COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv',  sep=',')
datos_global_recovered.head()
```

#### 4) countriesWorld

```python
datos_countriesWorld = pd.read_excel('countriesWorld_2.xlsx',  sheet_name='final')
datos_countriesWorld.head()
```

```python
datos_countriesWorld.shape
```

### --------------------------------------------------------------------------------------------------------------------------------------------

### UNIÓN DE DATASETS

```python
lista =list(datos_global_deaths.columns)[1:]
lista_countries =list(datos_countriesWorld.columns)
for i in range(len(lista_countries)-1):
    lista.append(lista_countries[i+1])
```

```python
full_death = pd.DataFrame(columns=lista)
#print(df)
myset = set(datos_global_deaths['Country/Region'])
i=0
for name in list(myset):
    full_death.loc[i,:3] = datos_global_deaths[(datos_global_deaths['Country/Region']==name)].iloc[0,1:4]
    full_death.loc[i,3:95] = (datos_global_deaths[(datos_global_deaths['Country/Region']==name)].iloc[:,4:]).sum(axis=0)
    full_death.loc[i,95:] = datos_countriesWorld[(datos_countriesWorld['Country/Region']==name)].iloc[0,1:]
    i += 1
    
```

```python
full_death[(full_death['Country/Region']=='Chile')]
```

### --------------------------------------------------------------------------------------------------------------------------------------------

### ELIMINACIÓN DATOS NaN O VACÍOS

```python
# BORRAR LOS NA
full_death.dropna(axis=0, inplace=True)
full_death = full_death.reset_index(drop=True)
```

```python
full_death.shape
```

```python
full_death.isnull().values.any()
```

### --------------------------------------------------------------------------------------------------------------------------------------------

### NORMALIZACION DE CASOS CONFIRMADOS

```python
full_death = full_death.sort_values('Country/Region')
```

```python
full_death.iloc[143]
```

```python
porcentajeMuertos = pd.DataFrame((100*full_death.iloc[:,94])/full_death.iloc[:,100],columns=['Porcentaje'])
porcentajeMuertos
```

```python
muertosNormalizados = pd.concat([full_death.iloc[:,0],porcentajeMuertos], axis=1)
muertosNormalizados
```

### --------------------------------------------------------------------------------------------------------------------------------------------

### TOP 30 CASOS MUERTOS

```python
top30_norm = muertosNormalizados.sort_values('Porcentaje',ascending=False).iloc[:29,:]
top30_norm
```

```python
muertosSinNormalizados = pd.concat([full_death.iloc[:,0],full_death.iloc[:,94]], axis=1)
top30_NO_norm = muertosSinNormalizados.sort_values('4/22/20',ascending=False).iloc[:29,:]
top30_NO_norm
```

### --------------------------------------------------------------------------------------------------------------------------------------------

### CURVAS DE TIEMPO: TOP 30 MUERTOS NORMALIZADOS

```python
top30_norm_lista = top30_norm['Country/Region'].unique()
top30_norm_lista[:10]
```

```python
lista_top30_norm =['Country/Region']
lista_fechas =list(datos_global_deaths.columns)[4:]
for i in range(len(lista_fechas)):
    lista_top30_norm.append(lista_fechas[i])
```

```python
full_death_top30_norm10 = pd.DataFrame(columns=lista_top30_norm)
full_death_top30_norm20 = pd.DataFrame(columns=lista_top30_norm)
#print(df)
myset_death10 = set(top30_norm_lista[:10])
myset_death20 = set(top30_norm_lista[10:])
i=0
for name in list(myset_death10):
    full_death_top30_norm10.loc[i,:1] = name
    full_death_top30_norm10.loc[i,1:] = (100*(datos_global_deaths[(datos_global_deaths['Country/Region']==name)].iloc[:,4:]).sum(axis=0))/datos_countriesWorld[(datos_countriesWorld['Country/Region']==name)].iloc[0,6]
    i += 1
for name in list(myset_death20):    
    full_death_top30_norm20.loc[i,:1] = name
    full_death_top30_norm20.loc[i,1:] = (100*(datos_global_deaths[(datos_global_deaths['Country/Region']==name)].iloc[:,4:]).sum(axis=0))/datos_countriesWorld[(datos_countriesWorld['Country/Region']==name)].iloc[0,6]
    i += 1
```

```python
df10 = full_death_top30_norm10.T
```

```python

header10 = df10.iloc[0]
df10 = df10[1:]
df10.columns = header10
df10
```

```python
df20 = full_death_top30_norm20.T
header20 = df20.iloc[0]
df20 = df20[1:]
df20.columns = header20
```

# ............................................
#### NORMALIZACIÓN DE DATOS: POR POBLACIÓN

```python
%matplotlib notebook
ax = df10.plot(style='*')
df20.plot(ax = ax, style='-')
plt.xlabel('Fecha (22/01/2020 - 20/04/2020)')
plt.ylabel('N° Muertos [%]')
plt.title('Casos muertos a la fecha, para TOP30 normalizado')
plt.legend(df10.columns)
plt.show()
```
